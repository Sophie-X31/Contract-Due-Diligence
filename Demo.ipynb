### Helper Functions

```python
import re
```

```python
def extract_text_from_pdf_info(data):
    texts = []
    for block in data:
        if block.get("type") == "text" and "text" in block:
            texts.append(block["text"])
    return texts

def split_clauses(text):
    pattern = r'\n(?=(?:\d+\.\s|[A-Z]\.\s))'
    clauses = re.split(pattern, text)
    return [c.strip() for c in clauses if len(c.strip()) > 80]

def split_numbered_clauses(text):
    pattern = r'(?<=\n)(?=(?:\d+(?:\.\d+)*\.?\s+|[A-Z]\.\s+|\([a-z]\)\s+))'
    clauses = [c.strip() for c in re.split(pattern, text) if len(c.strip()) > 50]
    return clauses
```

```python
def classify_clauses(clauses, tokenizer, model, batch_size=8, threshold=0.7):
    model.eval()

    results = []
    for i in tqdm(range(0, len(clauses), batch_size), desc="Classifying clauses"):
        batch = clauses[i:i+batch_size]

        inputs = tokenizer(
            batch,
            padding=True,
            truncation=True,
            return_tensors="pt"
        )

        with torch.no_grad():
            outputs = model(**inputs)
            probs = torch.nn.functional.softmax(outputs.logits, dim=1)
            preds = torch.argmax(probs, dim=1)

        for clause, pred, prob in zip(batch, preds, probs):
            confidence = prob[pred].item()
            label = label_encoder.inverse_transform([pred.item()])[0] if confidence >= threshold else None
            results.append({
                "clause": clause,
                "label": label,
                "confidence": confidence
            })

    return results
```

```python
def parse_risk_response(response):
    response = response.strip()

    # Format 1: Risk level + Explanation
    match = re.search(
        r'Risk\s*level\s*[:\-]?\s*(\w+).*?Explanation\s*[:\-]?\s*(.+)',
        response,
        flags=re.I | re.S
    )
    if match:
        risk_level = match.group(1).capitalize()
        explanation = match.group(2).strip()
        return risk_level, explanation

    # Format 2: Level: Explanation
    match = re.match(r'(\w+)\s*[:\-]\s*(.+)', response)
    if match:
        risk_level = match.group(1).capitalize()
        explanation = match.group(2).strip()
        return risk_level, explanation

    # Format 3: Risk Level: Low. Explanation
    match = re.search(r'Risk\s*Level\s*[:\-]?\s*(\w+)\.\s*(.+)', response, flags=re.I | re.S)
    if match:
        risk_level = match.group(1).capitalize()
        explanation = match.group(2).strip()
        return risk_level, explanation

    # Format 4: Risk Level: Low <space> Explanation
    match = re.search(r'Risk\s*Level\s*[:\-]?\s*(\w+)\s+(.+)', response, flags=re.I | re.S)
    if match:
        risk_level = match.group(1).capitalize()
        explanation = match.group(2).strip()
        return risk_level, explanation

    # Fallback
    return "Response Format Error", response
```

```python
def assess_risk(clauses_data, pipe, batch_size=8):
    risky_clauses = []
    for i in tqdm(range(0, len(clauses_data), batch_size), desc="Assessing risk"):
        batch = clauses_data[i:i+batch_size]

        for item in batch:
            # Prepare prompt for each clause
            clause_text = item["clause"]
            label = item["label"]
            confidence = item["confidence"]
            prompt = (
              f"Clause: {clause_text}\n"
              f"Classifier label: {label}, confidence: {confidence:.2f}\n\n"
              f"Analyze the risk associated with this clause and categorize it as No, Low, Medium, or High risk.\n"
              f"Provide a one-line explanation for any risk level above No risk.\n"
              f"Output format: 'Risk Level: explanation.'"
            )
            messages = [
                {"role": "system", "content": "You are a legal and investment expert performing contract due diligence."},
                {"role": "user", "content": prompt},
            ]

            # Send prompt to PHI-3 mini
            output = pipe(messages, **generation_args)
            response = output[0]['generated_text']

            # Parse response
            risk_level, explanation = parse_risk_response(response)
            risky_clauses.append({
                "clause": item["clause"],
                "risk_level": risk_level,
                "explanation": explanation,
                "label": label,
                "confidence": confidence
            })

    df = pd.DataFrame(risky_clauses)
    counts = df["risk_level"].value_counts().to_dict()
    counts = {k: counts.get(k, 0) for k in ["Low","Medium","High"]}
    return counts, df
```

### Widget

```python
import ipywidgets as widgets
from IPython.display import display
from glob import glob
import os
import json
import pandas as pd
import subprocess
```

```python
### File upload widget
upload = widgets.FileUpload(
    accept=".pdf",
    multiple=False,
    description="Upload PDF"
)
display(upload)

### Button to process PDF
process_btn = widgets.Button(description="Process PDF")
output_area = widgets.Output()
display(process_btn, output_area)

def on_process_clicked(b):
    with output_area:
        output_area.clear_output()
        if not upload.value:
            print("Please upload a PDF first!")
            return

        # Step 0: Save uploaded file
        uploaded_file = list(upload.value.values())[0]
        pdf_bytes = uploaded_file['content']
        pdf_name = uploaded_file['metadata']['name']
        pdf_path = f"/content/{pdf_name}"
        output_folder = f"/content/mineru_output"
        os.makedirs(output_folder, exist_ok=True)
        if os.path.exists(pdf_path):
            count = 0
            # print(f"PDF already exists at {pdf_path}, skipping upload...")
        else:
            with open(pdf_path, "wb") as f:
                f.write(pdf_bytes)
            print(f"Saved PDF as {pdf_path}")

            # Step 1: MinerU Convert PDF to JSON
            # use_gpu = torch.cuda.is_available()
            use_gpu = False
            cmd = [
                "mineru",
                "-p", pdf_path,
                "-o", output_folder
            ]
            if not use_gpu:
                cmd += ["-b", "pipeline"]
            print(f"Running MinerU... (GPU: {use_gpu})")
            result = subprocess.run(cmd, capture_output=True, text=True)
            # print("STDOUT:\n", result.stdout)
            # print("STDERR:\n", result.stderr)
            print("MinerU finished successfully.")

        # Step 1.5: Load JSON
        base_name = os.path.splitext(pdf_name)[0]
        json_path = None
        for root, dirs, files in os.walk(output_folder):
            for file in files:
                if file == f"{base_name}_content_list.json":
                    json_path = os.path.join(root, file)
                    break
        if not json_path:
            print("Could not find content_list.json!")
            return
        print(f"Found JSON: {json_path}")
        with open(json_path, "r") as f:
            data = json.load(f)

        # Step 2: Parse clauses
        texts = extract_text_from_pdf_info(data)
        full_text = "\n".join(texts)
        clauses = split_clauses(full_text)
        clauses = split_numbered_clauses(full_text)
        print(f"Extracted {len(clauses)} clauses")

        # Step 3: Run classifier
        classified_results = classify_clauses(clauses, bert_tokenizer, classifier)
        print("Classified clauses")

        # Step 4: Assess risk
        counts, df_risk = assess_risk(classified_results, pipe)
        print("Risk counts:", counts)

        # Step 5: Save results as CSV
        csv_path = f"/content/{pdf_name.replace('.pdf','_analysis.csv')}"
        df_risk.to_csv(csv_path, index=False)
        print(f"Analysis results saved to: {csv_path}")

process_btn.on_click(on_process_clicked)
```
